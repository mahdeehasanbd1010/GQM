```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
dataset = pd.read_csv("EncodedData.csv")
dataset.drop(columns=['Designation'], inplace=True)
#print(dataset.head)
dataset.dropna(inplace=True)
dataset.drop(columns=['Unnamed: 0'],inplace=True)
print(dataset.shape)
```

```python
data=dataset[['Syntax easy','Libraries Helpful','Community Support','Projects',
             'Used as',]]
cor=data.corr(method='spearman')
# setting the parameter values
annot = True
  
# plotting the heatmap
hm = sn.heatmap(data=cor,
                annot=annot)
  
# displaying the plotted heatmap
plt.show()
```
